```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_test.csv')
```

## 1. MinMaxScaler

### Question. train 데이터의 Age 컬럼을 MinMax 스케일링하여 얻은 값을 age_mm_scaling 컬럼에 추가한 후, train셋과 같은 기준으로 test 데이터의 Age를 스케일링하고 그 결과를 age_mm_scaling 컬럼에 추가하라.

####step1. 'MinMaxScaler'를 사용하기 위한 라이브러리를 import하기

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

#### step2. MinMaxScaler 객체 생성

```python
scaler = MinMaxScaler()
```

#### step3. train 데이터의 Age를 2차원 형태로 변환 후 이를 a1이라는 데이터로 저장

```python
a1 = train.Age.values.reshape(-1, 1)   ## a1 = train[['Age']]
a1
```

```python
a1 = train[['Age']]
a1
```

#### step4. MinMaxScaler 객체를 사용하여 데이터를 학습



```python
scaler.fit(a1)
```

#### step5. 학습된 MinMaxScaler 객체를 사용하여 'Age'열의 데이터를 변환하고, 'age_scaling'컬럼에 추가

```python
train['age_scaling'] = scaler.transform(a1)
```

#### step6. test 데이터에서도 'Age'컬럼을 추출하여 2차원 배열 형태로 변환하고, 학습된 MinMaxScaler 객체를 사용하여 test 데이터의 'Age'열을 변환 후 이를 'age_mm_scaling'컬럼에 추가

```python
test_age = test[['Age']]
test['age_mm_scaling'] = scaler.transform(test_age)
```

## 2. StandardScaler

### Question. train 데이터의 Age 컬럼을 sklearn 모듈을 활용하여 정규화 스케일링한 결과를 age_s_scaling 컬럼에 추가한 후, train셋과 같은 기준으로 test 데이터의 Age를 스케일링하여 age_s_scaling 컬럼에 추가하라.


#### step1. 'StandScaler'를 사용하기 위한 라이브러리를 import하기

```python
from sklearn.preprocessing import StandardScaler
```

#### step2. StandardScaler 객체 생성

```python
scaler2 = StandardScaler()
```

#### step3. train 데이터의 Age를 2차원 형태로 변환 후 이를 a1라는 데이터로 저장

```python
a1 = train[['Age']]
a1
```

#### step4. StandScaler 객체를 사용하여 데이터를 학습

```python
scaler2.fit(a1)
```

#### step5. 학습된  StandardScaler 객체를 사용하여 'Age'열의 데이터를 변환하고, 'age_s_scaling'컬럼에 추가

```python
train['age_s_scaling'] = scaler2.transform(a1)
```

#### step6. test 데이터에서도 'Age'컬럼을 추출하여 2차원 배열 형태로 변환하고, 학습된 StandardScaler 객체를 사용하여 test데이터의 'Age'열을 변환 후 'age_s_scaling'컬럼에 추가

```python
test_a1 = test[['Age']]
test['age_s_scaling'] = scaler2.transform(test_a1)
```
