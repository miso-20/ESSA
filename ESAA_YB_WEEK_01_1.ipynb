넘파이의 기반 데이터 타입 : ndarray ->  다차원 배열을 쉽게 생성하고 다양한 연산을 수행

array() 함수 :  다양한 인자를 입력받아서 ndarray로 변환하는 기능

 -  shape 변수 : 행과 열의 수를 튜플 형태로 가짐 -> ndarray 배열의 차원 알 수 있음

np.array() : 객체를 인자로 입력하면 ndarray로 반환

```python
array1 = np.array([1,2,3])
print('array1 type：',type(array1))
print('array1 array 형태:', array1.shape) ## 1차원 배열로 3개의 데이터를 가짐

array2 = np.array([[1,2,3],
                   [2,3,4]])
print('array2 type：',type(array2))
print('array2 array 형태:', array2.shape) ## 2차원 배열로 2행 3열 -> 6개의 데이터를 가짐

array3 = np.array([[1,2,3]])
print('array3 type：', type(array3))
print('array3 array 형태:', array3.shape) ## 2차원 배열로 1행 3열
```

   ### 2) ndarray의 데이터 타입

ndarray내의 데이터값 : 숫자 값, 문자열 값, 불 값 등이 모두 가능

- 숫자형의 경우 : int형, unsigned int형, float형, complex 타입

ndarray 내의 데이터 타입 : 같은 데이터 타입만 가능,  dtype 속성(int, float 등)으로 확인

```python
list1 = [1,2,3]
print(type(list1))
array1 = np.array(list1)
print(type(array1))
print(array1, array1.dtype)
```

ndarray 내의 데이터 타입이 다른 경우 : 데이터 크기가 더 큰 데이터 타입으로 형 변환

```python
list2 = [1, 2, 'test'] ## int형, 문자열: int형이 유니코드 문자열 값으로 변환
array2 = np.array(list2)
print(array2, array2.dtype) ## 숫자형 값 1, 2가 모두 문자열 값으로 변환

list3 = [1, 2, 3.0] ## int형, floaat형: float64형으로 변환
array3 = np.array(list3)
print(array3, array3.dtype)
```

ndarray 내 데이터값의 타입 변경 :  astype( ) 메서드를 이용 -> astype( )에 인자로 원하는 타입을 문자열로 지정
- 데이터 타입변경 : 메모리 절약할 때 이용

```python
array_int = np.array([1, 2, 3])
array_float = array_int.astype('float64') ## int32형 데이터를 float64로 변환
print(array_float, array_float.dtype)

array_int1= array_float.astype('int32') ## float64를 int32로 변환
print(array_int1, array_int1.dtype)

array_float1 = np.array([1.1, 2.1, 3.1 ])
array_int2= array_float1.astype('int32') ## float64를 int32로 변환
print(array_int2, array_int2.dtype)
```

### 4) ndarray의 차원과 크기를 변경하는 reshape()

reshape(): ndarray 를 특정 차원 및 크기로 변환

```python
array1 = np.arange(10)
print('array1: \n', array1)

array2 = array1.reshape(2, 5)
print('array2: \n', array2)

array3 = array1.reshape(5, 2)
print('array3：\n', array3)
```

```python
array1.reshape(4,3) ## 지정된 사이즈로 변경이 불가능하면 오류 발생
```

- reshape() -> -1을 인자로 사용하면 원래 ndarray와 호환되는 새로운 shape로 변환

```python
array1 = np.arange(10)
print(array1)
array2 = array1.reshape(-1,5) ## 고정된 5열에 맞는 행을 자동으로 새롭게 생성해 변환
print('array2 shape：', array2.shape)
array3 = array1.reshape(5,-1) ## 고정된 5행에 맞는 열을 자동으로 새롭게 생성해 변환
print('array3 shape：', array3.shape)
```

```python
array1 = np.arange(10) ## -1 사용해도 호환될 수 없는 형태는 변환X
array4 = array1.reshape(-1, 4)
```

 - reshape(—1,1) -> 여러 개의 행 & 1개의 열을 가진 ndarray로 변환

 - ndarray는 tolist() 메서드를 이용해 리스트 자료형으로 변환 가능

```python
array1 = np.arange(8)
array3d = array1.reshape((2, 2, 2))
print('array3d：\n', array3d.tolist())

# 3차원 ndarray를 2차원 ndarray로 변환
array5 = array3d.reshape(-1, 1)
print('array5：\n', array5.tolist())
print('array5 shape：', array5.shape)

# 1차원 ndarray를 2차원 ndarray로 변환
array6 = array1.reshape(-1, 1)
print('array6：\n', array6.tolist())
print('array6 shape：',array6.shape)
```

 (3) 팬시 인덱싱 : 리스트나 ndarray로 인덱스 집합을 지정하면 해당 위치의 인덱스에 해당하는 ndarray를 반환하는 인덱싱 방식

```python
array1d = np.arange(start=1, stop=10)
array2d = array1d.reshape(3, 3)

array3 = array2d[[0, 1], 2] ## 0,1번 행의 2번 열 선택
print('array2d[[0, 1], 2] => ', array3.tolist())

array4 = array2d[[0, 1], 0:2] ## 0,1번 행의 0,1번열 선택
print('array2d[[0,1], 0:2] => ', array4.tolist())

array5 = array2d[[0, 1]] ## 0,1번 행의 모든 열 선택
print('array2d[[0, 1]] => ', array5.tolist())
```

(4) 불린 인덱싱 : 조건 필터링과 검색을 동시에 할 수 있기 때문에 매우 자주 사용되 는 인덱싱 방식

- ndarray의 인덱스를 지정하는 [ ] 내에 조건문을 그대로 기재

불린 인덱싱 동작하는 단계

- Step 1: ndarray의 필터링 조건을 [ ] 안에 기재
- Step 2 : True 값에 해당하는 인덱스를 저장
- Step 3 : 저장된 인덱스 값으로 ndarray 조회

```python
array1d = np.arange(start=1, stop=10)
# [ ] 안에 array 1d > 5 Boolean indexing을 적용
array3 = array1d[array1d > 5]
print('array 1d > 5 불린 인덱싱 결과 값 :', array3)
```

```python
boolean_indexes = np.array([False, False, False, False, False, True, True, True, True])
array3 = array1d[boolean_indexes]
print('불린 인덱스로 필터링 결과 :', array3)
```

```python
indexes = np.array([5, 6, 7, 8]) ## 인덱스 5,6,7,8이니까 값은 6,7,8,9
array4 = array1d[indexes]
print('일반 인덱스로 필터링 결과 :', array4)
```

### 6) 행렬의 정렬 - sort()와 argsort()

(1) 행렬 정렬

np.sort() : 넘파이에서 sort()를 호출하는 방식, 원본 행렬을 변경하지 않고 정렬된 형태로 반환


ndarray.sort() : 행렬 자체에서 sort()를 호출하는 방식, 원본 행렬 자체를 정렬한 값으로 변환하며 반환 값은 None.

```python
org_array = np.array([ 3, 1, 9, 5])
print('원본 행렬:', org_array)

# np.sort( )로 정렬
sort_array1 = np.sort(org_array)
print ('np.sort( ) 호출 후 반환된 정렬 행렬:', sort_array1)
print('np.sort( ) 호출 후 원본 행렬:', org_array)

# ndarray.sort( )로 정렬
sort_array2 = org_array.sort()
print('org_array.sort( ) 호출 후 반환된 행렬:', sort_array2)
print('org_array.sort( ) 호출 후 원본 행렬:', org_array)
```

np.sort()나 ndarray.sort( ) 모두 기본적으로 **오름차순**으로 행렬 내 원소를 정렬

내림차순 정렬은 [：：-1]을 사용

```python
sort_array1_desc = np.sort(org_array)[::-1]
print ('내림차순으로 정렬:', sort_array1_desc)
```

행렬이 2차원 이상일 경우에 axis 축 값 설정을 통해 로우 방향, 또는 칼럼 방향으로 정렬할 수 있음

```python
array2d = np.array([[8, 12], [7, 1 ]])

sort_array2d_axis0 = np.sort(array2d, axis=0)
print('로우 방향으로 정렬:\n', sort_array2d_axis0)

sort_array2d_axis1 = np.sort(array2d, axis=1)
print('칼럼 방향으로 정렬:\n', sort_array2d_axis1)
```

(2) 정렬된 행렬의 인덱스를 반환하기

np.argsort() : 정렬 행렬의 원본 행렬 인덱스를 ndarray 형으로 반환

```python
org_array = np.array([ 3, 1, 9, 5])
sort_indices = np.argsort(org_array)
print(type(sort_indices))
print('행렬 정렬 시 원본 행렬의 인덱스:', sort_indices)
```

```python
org_array = np.array([ 3, 1, 9, 5])
sort_indices_desc = np.argsort(org_array)[::-1]
print('행렬 내림차순 정렬 시 원본 행렬의 인덱스:', sort_indices_desc)
```

argsort() : np.argsort(score_array)를 이용해 반환된 인덱스를 name_array에 팬시 인덱스로 적용해 추출할 수 있음

```python
import numpy as np

name_array = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
score_array= np.array([78, 95, 84, 98, 88])

sort_indices_asc = np.argsort(score_array)
print('성적 오름차순 정렬 시 score_array의 인덱스:', sort_indices_asc)
print('성적 오름차순으로 name_array의 이름 출력:', name_array[sort_indices_asc])
```

### 7) 선형대수 연산 - 행렬 내적과 전치 행렬 구하기

(1) 행렬 내적 (행렬 곱)

- np.dot() 을 이용해 계산이 가능


```python
A = np.array( [[1, 2, 3], [4, 5, 6]] )
B = np.array( [[7, 8], [9, 10], [11, 12]] )

dot_product = np.dot(A, B)
print('행렬 내적 결과:\n', dot_product)
```

(2) 전치 행렬
행과 열 위치를 교환한 원소로 구성한 행렬
- transpose()를 이용해 구할 수 있음

```python
A = np.array([[1, 2],[3, 4]])
transpose_mat = np.transpose(A)
print ('A 의 전치 행렬:\n', transpose_mat)
```
